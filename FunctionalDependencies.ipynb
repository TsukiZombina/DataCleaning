## Algoritmo de fuerza bruta para encontrar dependencias funcionales

El siguiente algoritmo de fuerza bruta determina las dependencias funcionales dados un data frame y dos columnas $c_{1}$ y $c_{2}$. 

```python
def get_fd(df, A, B):
    dependency = True
    for value in df[A].unique():
        # Get all entries that contain 'value' in A
        df_value = df[df[A] == value]
        
        #if len(df_value) > 1:
            # Check if all values in the column B are the same
        uniques = df_value[B].unique()
        #print('Uniques of column {0} under {1} = {2}'.format(B, A, value))
        #print(len(uniques))

        dependency = dependency and len(uniques) == 1

        if dependency:
            print('{0} = {1} -> {2} = {3}'.format(A, value, B, uniques[0]))
        

def search_fd(df):
    for B in df.columns:
        for A in df.columns:
            if B is not A:
                get_fd(df, A, B)
                
search_fd(df)
```

## Algoritmo para encontrar las dependencias funcionales de un conjunto

Al buscar posibles de pendencias funcionales, se pueden encontrar que existan conjuntos de atributos, tales que cumplen que A y B definen funcionalmente a C.

```python
def pair_fd(df, A, B, C):
    dependency = True
    for value in df[A].unique():
        # Get all entries that contain 'value' in C
        df_value = df[df[A] == value]
        
        # Check if all values in the column B are the same
        uniques = df_value[C].unique()
        dependency = dependency and len(uniques) == 1
        
    if dependency:
        for value in df[B].unique():
            # Get all entries that contain 'value' in B
            df_value = df[df[B] == value]

            # Check if all values in the column B are the same
            uniques = df_value[C].unique()
            dependency = dependency and len(uniques) == 1

    if dependency:
        print('{0}, {1} -> {2}'.format(A, B, C))
        
pair_fd(df, 'A', 'E', 'F')
```
